```python
# finished.jsonの読み込み
import json
json_open = open("finished.json", "r")
json_load = json.load(json_open)
print(len(json_load))
print(json_load[1])
```

```python
import pandas as pd

info = pd.read_csv("./final_product/data/chartmode_data.csv")
info.head()
```

```python
for i in range(len(json_load)):
    for j in range(len(info)):
        if json_load[i]["company_name"] == info["提出者名"][j]:
            json_load[i]["capital"] = info["資本金"][j]
            json_load[i]["found_year"] = info["設立年"][j]
            json_load[i]["listed_year"] = info["上場年"][j]
            json_load[i]["num_of_employees"] = info["従業員数"][j]
            json_load[i]["overtime"] = info["残業時間"][j]
            json_load[i]["paid_holiday"] = info["有休消化率"][j]
            json_load[i]["income"] = info["平均年収"][j]
            json_load[i]["age"] = info["平均年齢"][j]
```

```python
with open("test.json", "w") as f:
    json.dump(json_load, f , indent=4, ensure_ascii=False)
```
