```python
# finished.jsonの読み込み
import json
json_open = open("finished.json", "r")
json_load = json.load(json_open)
print(json_load[0])
print(len(json_load))
```

```python
import pandas as pd
basic_info = pd.read_csv("./final_product/data/chartmode_data.csv")
basic_info.head()
```

```python
for i in range(len(json_load)):
    name = json_load[i]["company_name"]
    for j in range(len(basic_info)):
        if basic_info["提出者名"][j] == name:
            json_load[i]["company_code"] = basic_info["株式コード"][j]
            json_load[i]["category_original"] = basic_info["提出者業種"][j]
            json_load[i]["category_new"] = basic_info["業種"][j]
            json_load[i]["capital"] = basic_info["資本金"][j]
            json_load[i]["found_year"] = basic_info["設立年"][j]
            json_load[i]["listed_year"] = basic_info["上場年"][j]
            json_load[i]["num_of_employees"] = basic_info["従業員数"][j]
            json_load[i]["overtime"] = basic_info["残業時間"][j]
            json_load[i]["paid_holiday"] = basic_info["有休消化率"][j]
            json_load[i]["income"] = basic_info["平均年収"][j]
            json_load[i]["age"] = basic_info["平均年齢"][j] 
            break
```
