```python
import pandas as pd
```

```python
data = pd.read_csv("./chartmode_data.csv")
data.head()
```

```python
category_count = {}
for i in range(len(data)):
    if(data["業種"][i] in category_count):
        category_count[data["業種"][i]] += 1
    else:
        category_count[data["業種"][i]] = 1
```
