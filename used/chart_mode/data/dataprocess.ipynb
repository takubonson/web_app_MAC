## ここで手動でいらないデータを削除する

```python
data = pd.read_csv("./all_data_tmp.csv")
data.head()
data.isnull().sum()
```

```python
# カテゴリーの種類
all_category = data["提出者業種"].values
big_category = data["業種"].values
categories = {}
for i in range(len(all_category)):
    if not big_category[i] in categories:
        categories[big_category[i]] = [all_category[i]]
    else:
        if not all_category[i] in categories[big_category[i]]:
            categories[big_category[i]].append(all_category[i])

# for category in all_category:
#     if category in categories:
#         continue
#     categories.append(category)
print(categories)
with open("category.txt", "w") as f:
    for category in categories:
        f.write('"' + category + '", ')
```

```python
# 平均年収の処理
for i in range(len(data)):
    salary = data["平均年収"][i]
    if salary == "None":
        data["平均年収"][i] = 0
    else:
        data["平均年収"][i] = int(data["平均年収"][i].replace("千円", "").replace(",", ""))
```

```python
# 平均年齢の処理
for i in range(len(data)):
    age = data["平均年齢"][i]
    if age == "None":
        data["平均年齢"][i] = 0
    else:
        data["平均年齢"][i] = float(age.replace("歳", ""))
```

```python
# 従業員数の処理
for i in range(len(data)):
    num_of_people = data["従業員数(連結)"][i]
    if num_of_people == "None":
        data["従業員数(連結)"][i] = 0
    else:
        data["従業員数(連結)"][i] = int(data["従業員数(連結)"][i].replace("人", "").replace(",", ""))
```

```python
data.head()
```

```python
data["設立年"] = ""
data["上場年"] = ""
for i in range(len(data)):
    date_found = data["設立年月日"][i]
    date_jojo = data["上場年月日"][i]
    
    if date_found == "None":
        data["設立年"][i] = "None"
    else:
        data["設立年"][i], _ = date_found.split("年")
        data["設立年"][i] = int(data["設立年"][i])
        
    if date_jojo == "None":
        data["上場年"][i] = "None"
    else:
        data["上場年"][i], _ = date_jojo.split("年")
        data["上場年"][i] = int(data["上場年"][i])

data = data.drop("設立年月日",axis=1)
data = data.drop("上場年月日",axis=1)
```

```python
data.head()
```

```python
# 時価総額の処理
for i in range(len(data)):
    data["時価総額"][i] = float(data["時価総額"][i].replace("$", "").replace("B",""))
```
